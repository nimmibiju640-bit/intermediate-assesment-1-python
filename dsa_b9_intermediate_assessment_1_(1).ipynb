## Challenge 1: Word Frequency Analyzer
Write a function `top_k_words(filename: str, k: int) -> list` that returns the top k frequent words.
- Ignore case and punctuation.
- Sort by frequency, then alphabetically.


```python
import re
from collections import Counter

def top_k_words(filename: str, k: int) -> list:
    with open(filename, "r") as f:
        text = f.read().lower()

    # Remove punctuation and extract words
    words = re.findall(r'\b\w+\b', text)

    # Count frequency
    freq = Counter(words)

    # Sort by (-frequency, word)
    sorted_words = sorted(freq.items(), key=lambda x: (-x[1], x[0]))

    # Return top k words
    return [word for word, count in sorted_words[:k]]


# Sample Test
with open("sample.txt", "w") as f:
    f.write("Data science is evolving. Python is versatile. Learning never stops.")

print(top_k_words("sample.txt", 3))

# Sample Test (create a sample file)
with open("sample.txt", "w") as f:
    f.write("Data science is evolving. Python is versatile. Learning never stops.")

print(top_k_words("sample.txt", 3))  #expected ["is", "data", "evolving"]
```

## Challenge 2 : Analyze daily expenses and extract insights.
User enters:

Number of transactions
Category and amount

Food 200
Travel 500
Food 300
Shopping 700

### Output
* Total expense
* Highest expense category
* Category-wise total

```python
def analyze_expenses():
    n = int(input("Enter number of transactions: "))

    expenses = {}

    for _ in range(n):
        try:
            parts = input().split()

            # Ensure valid input format
            if len(parts) < 2:
                print("Invalid input, skipping...")
                continue

            category = " ".join(parts[:-1])  # handles multi-word categories
            amount = float(parts[-1])

            # Ignore negative values
            if amount < 0:
                print("Negative value ignored")
                continue

            expenses[category] = expenses.get(category, 0) + amount

        except:
            print("Invalid input, skipping...")

    # Calculate total
    total = sum(expenses.values())

    # Highest category
    if expenses:
        highest_category = max(expenses, key=expenses.get)
        max_expense = expenses[highest_category]
    else:
        highest_category = None
        max_expense = 0

    print("Total Expense:", total)
    print(f"Highest Category: {highest_category} ({max_expense})")
    print("Category-wise:", expenses)


analyze_expenses()
```

### Challenge 3: Data Cleaner (Mini Pandas Logic Without Pandas)

10 ,20 ,abc ,30 ,-5 , 40

* Remove invalid values
* Remove negative values


Show:
* Clean list
* Average
* Max value

```python
def clean_data():
    n = int(input("Enter number of values: "))

    data = []

    for _ in range(n):
        value = input()

        try:
            num = float(value)

            # Ignore negative values
            if num >= 0:
                data.append(num)

        except:
            # Ignore invalid values
            continue

    # Calculate average and max
    if len(data) > 0:
        avg = sum(data) / len(data)
        max_val = max(data)
    else:
        avg = 0
        max_val = None

    print("Clean Data:", data)
    print("Average:", avg)
    print("Max:", max_val)


clean_data()
```
